```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('http://media.sundog-soft.com/SelfDriving/FuelEfficiency.csv')
df.head()
```

```python
df2 = df[['Cylinders','CityMPG','HwyMPG','CombMPG']]
df2.head()
```

```python
import seaborn as sns
sns.pairplot(df2, hue='Cylinders',palette='Set1')
```

```python
sns.lmplot(df,x='Eng Displ',y='CombMPG')
```

```python
plt.figure(figsize=(12,3))
ax=sns.boxplot(df,x='Mfr Name',y='CombMPG',palette='Set1')
ax.set_xticklabels(ax.get_xticklabels(),rotation=90)
```

```python
plt.figure(figsize=(12,3))
ax=sns.swarmplot(df,x='Mfr Name',y='CombMPG',palette='Set1',size=2)
ax.set_xticklabels(ax.get_xticklabels(),rotation=90)
```
