```python
from sklearn.datasets import load_iris
iris = load_iris()
numSamples, numFeatures = iris.data.shape
print(numSamples)
print(numFeatures)
print(list(iris.target_names))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.2, random_state=0)
```

```python
import xgboost as xgb
train = xgb.DMatrix(X_train, label=y_train)
test = xgb.DMatrix(X_test, label=y_test)
```

```python
param = {
    'max_depth': 4,
    'eta': 0.3,
    'objective': 'multi:softmax',
    'num_class': 3
}
epochs = 10
```

```python
model = xgb.train(param, train, epochs)
```

```python
predictions = model.predict(test)
```

```python
print(predictions)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, predictions)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions))
```

```python
import os
import pandas as pd
rating_path = os.getcwd() +'/ml-course-materials/ml-100k/u.data'
movie_path = os.getcwd() +'/ml-course-materials/ml-100k/u.item'
```

```python
r_cols = ['user_id', 'movie_id', 'rating']
ratings = pd.read_csv(rating_path, sep='\t', names=r_cols, usecols=range(3))

m_cols = ['movie_id', 'title']
movies = pd.read_csv(movie_path, sep='|', names=m_cols, usecols=range(2), encoding="ISO-8859-1")
```

```python
movie_ratings = pd.merge(movies, ratings)
```

```python
movie_ratings.head()
```

```python
movie_ratings.info()
```

```python
movie_ratings.groupby('title')['rating'].mean().sort_values(ascending=False).head()
```

```python
pivot_movie_ratings = pd.pivot_table(movie_ratings, index='user_id', columns='title', values='rating')
pivot_movie_ratings.head()
```

```python
starWarsRatings = pivot_movie_ratings['Star Wars (1977)']
starWarsRatings.head()
```

```python
similarMovies = pivot_movie_ratings.corrwith(starWarsRatings)
similarMovies = similarMovies.dropna()
df = pd.DataFrame(similarMovies)
df.head(10)
```

```python
similarMovies.sort_values(ascending=False)
```

```python
import numpy as np
movieStats = movie_ratings.groupby('title').agg({'rating': [np.size, np.mean]})
movieStats.head()
```

```python
popularMovies = movieStats['rating']['size'] >= 100
movieStats[popularMovies].sort_values([('rating', 'mean')], ascending=False).head()
```

```python
df = movieStats[popularMovies].droplevel(0,axis=1).join(pd.DataFrame(similarMovies, columns=['similarity']))
```

```python
df.sort_values(by='similarity', ascending=False).head()
```

```python
#Item-based collaborative filterings
```

```python
userRatings = movie_ratings.pivot_table(index=['user_id'], columns=['title'], values='rating')
userRatings.head()
```

```python
corrMatrix = userRatings.corr()
corrMatrix.head()
```

```python
corrMatrix = userRatings.corr(method='pearson', min_periods=100)
corrMatrix.head()
```

```python
myRatings = userRatings.loc[0].dropna()
myRatings
```

```python
simCandidates = pd.Series()
for i in range(0, len(myRatings.index)):
    print("Adding sims for " + myRatings.index[i] + "...")
    sims = corrMatrix[myRatings.index[i]].dropna()
    sims = sims.map(lambda x: x * myRatings.iloc[i])
    # print(sims)
    simCandidates = pd.concat([simCandidates, sims])
print("sorting...")
simCandidates.sort_values(inplace = True, ascending = False)
print(simCandidates.head(10))
```

```python
simCandidates = simCandidates.groupby(simCandidates.index).sum()
```

```python
simCandidates.sort_values(inplace = True, ascending = False)
simCandidates.head(10)
```

```python
filteredSims = simCandidates.drop(myRatings.index)
filteredSims.head(10)
```
